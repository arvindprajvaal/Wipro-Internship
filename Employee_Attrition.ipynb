```python
combined_df['Gender'] = combined_df['Gender'].map({'Male':0,'Female':1})
combined_df['Job Role'] = combined_df['Job Role'].map({'Education':1,'Media':2,'Healthcare':3,'Technology':4,'Finance':5})
combined_df['Work-Life Balance'] = combined_df['Work-Life Balance'].map({'Poor':0,'Fair':1,'Good':2,'Excellent':3})
combined_df['Job Satisfaction'] = combined_df['Job Satisfaction'].map({'Low':0,'Medium':1,'High':2,'Very High':3})
combined_df['Performance Rating'] = combined_df['Performance Rating'].map({'Low':0,'Below Average':1,'Average':2,'High':3})
combined_df['Overtime'] = combined_df['Overtime'].map({'No':0,'Yes':1})
combined_df['Education Level']= combined_df['Education Level'].map({"High School":0,"Bachelor’s Degree":1,"Master’s Degree":2,"Associate Degree":3,"PhD":4})
combined_df['Marital Status']= combined_df['Marital Status'].map({'Single':0,'Married':1,'Divorced':2})
combined_df['Job Level']= combined_df['Job Level'].map({'Entry':0,'Mid':1,'Senior':2})
combined_df['Company Size']= combined_df['Company Size'].map({'Small':0,'Medium':1,'Large':2})
combined_df['Remote Work']= combined_df['Remote Work'].map({'No':0,'Yes':1})
combined_df['Leadership Opportunities']= combined_df['Leadership Opportunities'].map({'No':0,'Yes':1})
combined_df['Innovation Opportunities']= combined_df['Innovation Opportunities'].map({'No':0,'Yes':1})
combined_df['Company Reputation']= combined_df['Company Reputation'].map({'Poor':0,'Fair':1,'Good':2,'Excellent':3})
combined_df['Employee Recognition'] = combined_df['Employee Recognition'].map({'Low':0,'Medium':1,'High':2,'Very High':3})
combined_df['Attrition'] = combined_df['Attrition'].map({'Left':0,'Stayed':1})

#Encoding Categorical Variables to Numerical valuses using map() function
```

```python
numerical_cols = combined_df.select_dtypes(include=np.int64).columns


z_scores = np.abs(stats.zscore(combined_df[numerical_cols]))


threshold_z = 3


outlier_rows = (z_scores > threshold_z).any(axis=1)


combined_df_cleaned = combined_df[~outlier_rows].reset_index(drop=True)

print("Original DataFrame Shape:", combined_df.shape)
print("DataFrame Shape after Removing Outliers:", combined_df_cleaned.shape)

# Removing the outliers in the dataset(with a threshold of 3) using Z Score method and printing the shape comparison
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#Importing Standard Scaler for scaling values
```

```python
from sklearn.model_selection import train_test_split

X=combined_df_cleaned.drop(columns=['Attrition','Employee ID'],axis=1)
y=combined_df_cleaned['Attrition']

#Forming the X and y variables for train-test splitting
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)

#Splitting the dataset into X_train,X_test, y train and y test in the ratio 80:20 and scaling the X train and test values
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

#Importing the models along with classification report to evaluate the models
```

```python
n=int(input('Enter neighbours for KNN:'))
knn= KNeighborsClassifier(n_neighbors=n,metric='minkowski')
knn.fit(X_train_scaled,y_train)
knn_pred= knn.predict(X_test_scaled)

print(classification_report(y_test,knn_pred))

#Fitting the training data in the KNN Model,predicting and then evaluating the model using Classification report
```

```python
nb=GaussianNB()
nb.fit(X_train_scaled,y_train)
nb_pred=nb.predict(X_test_scaled)

print(classification_report(y_test,nb_pred))

#Fitting the training data in the Naive Bayes Model,predicting and then evaluating the model using Classification report
```

```python
dec_tree= DecisionTreeClassifier(criterion='entropy',max_depth=12,random_state=40)
dec_tree.fit(X_train_scaled,y_train)
dec_tree_pred= dec_tree.predict(X_test_scaled)

print(classification_report(y_test,dec_tree_pred))

#Fitting the training data in the Decision Tree Model,predicting and then evaluating the model using Classification report
```

```python
random_forest= RandomForestClassifier(n_estimators=130,criterion='entropy',max_depth=20,random_state=40)
random_forest.fit(X_train_scaled,y_train)
random_forest_pred= random_forest.predict(X_test_scaled)

print(classification_report(y_test,random_forest_pred))

#Fitting the training data in the Random Forest Model,predicting and then evaluating the model using Classification report
```

```python
logreg= LogisticRegression()
logreg.fit(X_train_scaled,y_train)
logreg_pred= logreg.predict(X_test_scaled)

print(classification_report(y_test,logreg_pred))

#Fitting the training data in the Logistic Regression Model,predicting and then evaluating the model using Classification report
```

```python
grad_boost= GradientBoostingClassifier(n_estimators=100,learning_rate=0.1,max_depth=10,random_state=40)
grad_boost.fit(X_train_scaled,y_train)
grad_boost_pred= grad_boost.predict(X_test_scaled)

print(classification_report(y_test,grad_boost_pred))

#Fitting the training data in the Gradient Boosting Classifier Model,predicting and then evaluating the model using Classification report
```

```python
rf= RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train_scaled,y_train)
importances= rf.feature_importances_
```

```python
from sklearn.ensemble import AdaBoostClassifier, VotingClassifier,BaggingClassifier
import xgboost as xgb
import lightgbm as lgb
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score,recall_score,f1_score

#Importing Ensemble Models
```

```python
def evaluate_model(name, model, X_train, y_train, X_val, y_val):
    model.fit(X_train, y_train)
    preds = model.predict(X_val)

    print(f"\n📊 {name} Results")
    print("-" * 40)
    print("Accuracy:", accuracy_score(y_val, preds))
    print("Precision:", precision_score(y_val, preds))
    print("Recall:", recall_score(y_val, preds))
    print("F1 Score:", f1_score(y_val, preds))
    print("Confusion Matrix:\n", confusion_matrix(y_val, preds))
    print("Classification Report:\n", classification_report(y_val, preds))
```

```python
ensemble_models = {
    "AdaBoost": AdaBoostClassifier(n_estimators=150, random_state=42,learning_rate=1),
    "XGBoost": xgb.XGBClassifier(n_estimators=200, learning_rate=0.1, random_state=42, use_label_encoder=False, eval_metric='logloss',max_depth=3),
    "LightGBM": lgb.LGBMClassifier(n_estimators=50, learning_rate=0.1, random_state=42,num_leaves=31),
    "Bagging": BaggingClassifier(n_estimators=100, random_state=42),

    "Voting Ensemble (Soft)": VotingClassifier(
        estimators=[
            ('lr', LogisticRegression(max_iter=1000)),
            ('rf', RandomForestClassifier(n_estimators=100)),
            ('gb', GradientBoostingClassifier(n_estimators=100))
        ],
        voting='soft',weights=[1,2,1]
    )
}

#Training and evaluating the models
```

```python
for name, model in ensemble_models.items():
    evaluate_model(name, model, X_train_scaled, y_train, X_test_scaled, y_test)
```
